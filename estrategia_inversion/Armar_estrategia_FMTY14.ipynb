Empezamos llamando a la biblioteca que nos permite manejar las fechas en **Python**, esta se llama ```datetime```.
Para más información sobre la biblioteca puede consultar [esta página](https://www.w3schools.com/python/python_datetime.asp)

````python
from datetime import datetime #para manejar las fechas
````

Para obtener datos como el año, el mes o el día de la semana se necesita un objeto de tipo ```datetime```.
Cuando importamos los datos a **Python** la biblioteca ```pandas``` no distingue si la entrada es una fecha o una palabra, así que las fechas son exportadas con el formato ```string``` y lo vamos a transformar a un objeto de tipo ```datetime```.
Voy a crear una columna con el mes para separar la tabla.
Y una columna con el año que nos va a ser útil después.

Para transformar los datos de la primera columna en una fecha utilizamos  la función ```strptime```.
Para que  ```strptime``` transforme un ```string``` necesita esta entrada; pero existen muchas maneras de escribir una fecha, para distinguir cómo se escribió la fecha esta biblioteca tiene un formato.
El formato de esta columna empieza con el año completo, en la biblioteca se distingue por ```%Y```; luego el número del mes, se distingue por ```%m```; la última parte es contiene el día del mes, que se distingue por ```%d```.
Este método también nos regresa elementos de la fecha.

Empezamos creando dos columnas vacías.
Para llenar cada una de las entradas utilizamos un ciclo ```for```, empezamos creando una variable llamada ```s``` la fecha de la columna y usamos el método ```strptime``` para obtener el mes y el año.

````python
datos_fmty["Mes"] = None
datos_fmty["Año"] = None
for i in range(0,len(datos_fmty["Date"])):
    s = datetime.strptime(datos_fmty["Date"][i], '%Y-%m-%d')
    datos_fmty["Mes"][i] = s.strftime("%m")
    datos_fmty["Año"][i] = s.strftime("%Y")
````

La teoría para esta parte es muy extensa para este artículo, para entender mejor recomiendo leer los siguientes artículos [primer artículo](https://machinelearningmastery.com/decompose-time-series-data-trend-seasonality/), [segundo artículo](https://www.section.io/engineering-education/time-series-decomposition-in-python/) y [tercer artículo](https://towardsdatascience.com/time-series-decomposition-in-python-8acac385a5b2).

En resumen las funciones se pueden dividir en tres partes distintas:

La primera parte es la  es la **tendencia**.
Esta nos indica el promedio de la función en el tiempo, sí este es creciente concluimos que durante este tiempo el precio de las fibras se ha incrementado, haciendo a esta fibra una buena inversión para el futuro.

La segunda componente es la **periodicidad**.
Esta es la componente que vamos a utilizar para el análisis, cuando esta baje nos indica que el precio está bajando.

La última componente es el **ruido**.
Esta es totalmente impredecible, esto hace imposible saber la fecha exacta de cuándo va a bajar el precio de una acción.

Ahora tenemos que tomar en cuenta las tres componentes en conjunto.
Estaremos estudiando la **periodicidad** y para nuestra estrategia elegimos el día donde la periodicidad nos diga que el precio baje.
Pero tenemos que tomar en cuenta la **tendencia**, si la tendencia va aumentando el precio va aumentando, si la periodicidad baja después este efecto puede cancelarse.
Con la componente final no existe ninguna esperanza, ya que puede hacer que el día que elegimos el precio se dispare u otro día sea perfecto para comprar.

Desafortunadamente la situación en el mercado está cambiando continuamente, por ejemplo, el que aparecieran los servicios de streaming afectó los clubes de video; la pandemia afectó varias industrias cambiando el promedio de estas.

<font color='orange' size="6">Análisis anual </font>

Para darse una idea de cómo se comporta una acción es una buena idea saber cómo se comporta año con año, así que empezamos realizando un análisis de la fibra anual.
El tiempo en que debemos realizar el análisis va a afectar el código y los resultados.
Tenemos cuatro columnas para realizar el análisis, así que tenemos que decidir qué columnas son importantes para el análisis y cuáles no.
Los precios de inicio y fin nos pueden dar información sobre los días donde ocurrieron hechos importantes que pueden afectar el precio de la fibra, pero en este momento queremos tener una idea general del precio que tenemos que pagar; lo mejor es comparar los precios máximos **(High)** y mínimos **(Low)** para tener una idea del rango a pagar.

Como mencionábamos anteriormente la función ```seasonal_decompose``` va a ser la estrella en esta descomposición, por la manera en que llamamos a la función, esta va a ser llamada con el código ```sm```.
Nosotros tenemos vectores con los precios, al aplicar la función  ```seasonal_decompose```  creamos objetos nuevos de tipo ```seasonal.DecomposeResult```.
Este objeto tiene las tres componentes de los datos con los precios indicados.

Ahora hablaremos a detalle del método utilizado para la descomposición.
El primero es nombrar las variables de tipo ```seasonal.DecomposeResult``` para diferenciarlas, como van a darnos la descomposición las llamamos ```descomposicion_``` y aclaramos si es del mínimo o el máximo.

Los parámetros que hay que especificar el primero son los datos cuyas componentes queremos separar, para esto se utiliza el código ```x=``` igualando a los datos deseados.
El siguiente es especificar el tipo de modelo que vamos a utilizar, no hable de los dos tipos de modelos para realizar este análisis, para eso es mejor leer los artículos que dejé, pero esta entrada se especifica con ```model =```; empezamos el análisis para el modelo aditivo, así que igualamos eso a ```additive```.
El último dato importante que hay que especificar es la frecuencia que nos importa, una función puede ser descompuesta en muchas frecuencias, pero en nuestro análisis sólo nos importa una frecuencia.
Estamos haciendo un análisis anual, así que la frecuencia que deseamos son 12 por el número de meses, para indicar esto usamos el código ```freq=12```.

````python
descomposicion_min = sm(x=datos_fmty["Low"], model='additive', freq=12)
descomposicion_max = sm(x=datos_fmty["High"], model='additive', freq=12)
````

Para obtener las componentes aplicamos los métodos correspondientes a los objetos ```descomposicion_```.
Para distinguir si las componentes son el máximo o el mínimo se nombran con ```max``` y ```min```.

La componente de la **tendencia** se obtiene utilizando con el método ```trend```.

La componente **periodicidad** nos da las ```seasons``` con el método ```seasonal```.

Finalmente guardamos el **ruido** lo obtenemos con el método ```resid```.

````python
tendencia_min = descomposicion_min.trend
periodo_min = descomposicion_min.seasonal
ruido_min = descomposicion_min.resid

tendencia_max = descomposicion_max.trend
periodo_max = descomposicion_max.seasonal
ruido_max = descomposicion_max.resid
````

##  <font color='blue' size="7">Comprobando los resultados </font>

<font color='red'>Habilidades utilizadas:

 - Combinar los datos y crear gráficas para realizar una comparación de estos.</font>

El siguiente paso es comprobar si esta estrategia es eficiente.
Para esto vamos a simular que compramos una fibra Monterrey mensualmente usando la estrategia deseada, después simulamos comprar una fibra Monterrey mensualmente utilizando otras estrategias.
Para que el análisis sea sencillo armamos gráficas con el dinero que se ha invertido en ese momento para ver qué estrategia es más rentable.
No vamos a tomar en cuenta los dividendos.

Vamos a escoger qué estrategias vamos a comparar.
La mayoría de días se encuentran al final del mes, pero existen tendencias que son bajas al principio de mes, así que la primera estrategia con la que comparamos todo se encuentra la principio del mes, esta la llamamos ```principio```.
Otra estrategia es comprar las cosas a la mitad del mes, esta estrategia la llamamos ```mitad```.
Existen muchas maneras de escoger nuestra estrategia, pero vamos a comprar las acciones en un momento aleatorio para comparar si es necesario realizar un análisis tan detallado, a esta estrategia la llamamos ```aleatorio```.
A nuestra estrategia la llamamos simplemente ```estrategia```.

Empezamos creando arreglos para guardar los precios de las fibras según la estrategia deseada; vamos a usar el precio máximo y el precio mínimo para tener el rango de precios.
Los arreglos para el mínimo tienen el nombre ```_low``` y los máximos ```_max```.
Para crear arreglos de números usamos la biblioteca ```numpy```, para entender mejor esta biblioteca recomiendo leer [numpy](https://aprendeconalf.es/docencia/python/manual/numpy/).

Podemos crear una lista e ir agregando los valores, pero tendríamos que utilizar más memoria, lo que ralentiza el programa.

````python
import numpy as np #varias operaciones matemáticas
````

Creamos vectores que guardan el precio máximo y el precio mínimo del día de la estrategia, son tres años, que equivalen a 36 meses.
Por la base de datos escogida los últimos precios de noviembre y diciembre se van a escoger del 2019.

````python
principio_low = np.zeros(36) #precios menores
mitad_low = np.zeros(36)
aleatorio_low = np.zeros(36)
estrategia_low = np.zeros(36)

principio_high = np.zeros(36) #precios mayores
mitad_high = np.zeros(36)
aleatorio_high = np.zeros(36)
estrategia_high = np.zeros(36)
````

Creamos una columna para indicar el mes y otra columna para indicar el año, para tener el día correcto vamos a filtrar por los valores de estas dos columnas.
Para elegir los meses adecuados tenemos que separar los datos por año y mes, el método ```loc``` puede elegir renglones usando más de un criterio.

Para ahorrar código vamos a llenar los valores de los primeros 10 meses usando dos ciclos ```for```.
El primer ciclo ```for``` es para poner el año, como cada año tiene doce meses este ciclo va recorriendo el contador doce lugares, esto se logra haciendo ```i*12```; el rango de este ciclo es de 0 a 2.
El segundo ciclo ```for``` va indicando el mes, este empieza llenando de 0 a 9, después de 12 a 21 y finalmente de 24 a 33; el rango de este ciclo es de 0 a 9.

Empezamos creando vectores con los valores que van a indicar los primeros 10 meses llamado ```vector_meses```.
Otro vector con los valores del año usando ```vector_year```.
Y el último vector con el número de día en que vamos a comprar la fibra, con el nombre ```vector_estrategia```; como los meses no tienen el mismo número de días, para algunos meses se compró antes.

````python
vector_meses = ["01","02","03","04","05","06","07","08","09","10"]
vector_year = ["2020","2021","2022"]
vector_estrategia = [20,18,20,15,14,22,21,21,21,21,12,19]
````

Para llenar el vector aleatorio necesitamos llamar a la biblioteca ```random``` que nos va a dar acceso a la función ```randrange```.
Esta función nos regresa un entero aleatorio en un rango especificado, si tenemos un vector ```vec``` podemos elegir un elemento aleatorio de este vector escogemos un entero en el rango $[0, len(vec)-1]$ y utilizamos ese número para obtener el elemento, para escoger el número usamos ```randrange(len(vec))```.

Para elegir el día de en medio vamos a dividir los días del mes entre dos, como no todos los meses tienen un número de días par utilizo la función ```floor``` que nos da el entero que se encuentra por debajo del número.
Esta función se encuentra en la biblioteca ```math```.

````python
import random
import math
````

Ahora colocamos los ciclos para los primeros diez meses.
El ciclo ```for``` necesita una variable para distinguir cómo se utiliza, para el primer ciclo usamos la letra ```i```, para el segundo ciclo usamos la letra ```j```.
Para poner el rango utilizamos la función ```range(x,y)```; en este caso ```x``` es el primer valor que toma ```i```, e ```y``` es el último valor más uno que toma ```i```.

````python
for i in range(0,3):
    for j in range(0,10):
        #toamos la parte del DataFrame que nos interesa
        conjunto=datos_fmty.loc[(datos_fmty['Mes'] == vector_meses[j]) & (datos_fmty['Año'] == vector_year[i])].reset_index()
        #generamos los numeros que vamos a utilizar
        tamano = len(conjunto["Low"])
        numero = random.randrange(tamano)
        mitad = int(math.floor(tamano/2))
        s = vector_estrategia[j]-1
        #asignamos los valores
        principio_low[j+i*12] = conjunto["Low"][0]
        principio_high[j+i*12] = conjunto["High"][0]
        aleatorio_low[j+i*12] = conjunto["Low"][numero]
        aleatorio_high[j+i*12] = conjunto["High"][numero]
        estrategia_low[j+i*12] = conjunto["Low"][s]
        estrategia_high[j+i*12] = conjunto["High"][s]
        mitad_low[j+i*12] = conjunto["Low"][mitad]
        mitad_high[j+i*12] = conjunto["High"][mitad]
````

Podemos combinar los códigos necesarios para los meses de noviembre y diciembre, pero los separamos para entender mejor qué pasa.

Por la complejidad del mercado no podemos dar el día donde los precios serán menores, así que nosotros queremos saber al final de los tres años cuál es la estrategia que menos dinero costó, así que vamos a sumar mes a mes el costo de comprar en cada estrategia y al final tendremos una respuesta.

Empezamos con los vectores que tendran el costo total.

````python
principio_low_sum = np.zeros(36)
mitad_low_sum = np.zeros(36)
aleatorio_low_sum = np.zeros(36)
estrategia_low_sum = np.zeros(36)
principio_high_sum = np.zeros(36)
mitad_high_sum = np.zeros(36)
aleatorio_high_sum = np.zeros(36)
estrategia_high_sum = np.zeros(36)
````

Definimos el valor de la primera entrada del vector, que es simplemente el precio del primer día de la estrategia:

````python
principio_low_sum[0] = principio_low[0]
mitad_low_sum[0] = mitad_low[0]
aleatorio_low_sum[0] = mitad_low[0]
estrategia_low_sum[0] = estrategia_low[0]
principio_high_sum[0] = principio_high[0]
mitad_high_sum[0] = mitad_high[0]
aleatorio_high_sum[0] = aleatorio_high[0]
estrategia_high_sum[0] = estrategia_high[0]
````

Ahora utilizamos un ciclo ```for``` para llenar el resto de los datos.
Este será la suma del hasta el mes anterior, más el nuevo precio al que adquirimos la fibra:

````python
for i in range(1,36):
    principio_low_sum[i] = principio_low[i] + principio_low_sum[i-1]
    mitad_low_sum[i] = mitad_low[i] + mitad_low_sum[i-1]
    aleatorio_low_sum[i] = aleatorio_low[i] + aleatorio_low_sum[i-1]
    estrategia_low_sum[i] = estrategia_low[i] + estrategia_low_sum[i-1]
    principio_high_sum[i] = principio_high[i] + principio_high_sum[i-1]
    mitad_high_sum[i] = mitad_high[i] + mitad_high_sum[i-1]
    aleatorio_high_sum[i] = aleatorio_high[i] + aleatorio_high_sum[i-1]
    estrategia_high_sum[i] = estrategia_high[i] + estrategia_high_sum[i-1]
````

````python
plt.figure(figsize=(12, 6))
plt.xlabel('Mes', fontdict = font2)
plt.ylabel('Gastos', fontdict = font2)
plt.title('Suma de precios mínimos', fontdict = font1)
plt.plot(principio_low_sum, label = "comprar al principio")
plt.plot(aleatorio_low_sum, label = "si lo compramos en un momento cualquiera")
plt.plot(mitad_low_sum, label = "a la mitad del mes")
plt.plot(estrategia_low_sum, label = "utilizando la estrategia")

plt.ylim([12,410])
plt.legend()
plt.grid(True)
plt.show()
````
